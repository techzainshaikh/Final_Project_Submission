```python
# Read data from file and handle null value
import pandas as pd

df = pd.read_csv('orders.csv', na_values=['Not Available','unknown'])
df['Ship Mode'].unique()
```

```python
# Derive new column discount, sale price, profit 
# df['discount']=df['list_price']*df['discount_percent']*.01 # *.01 ya divide by 100 same thing
# df['sales_price']=df['list_price']-df['discount']
df['profit']=df['sales_price']-df['cost_price']
df.head(5)
```

```python
# Convert order date data type object to datetime

df['order_date']=pd.to_datetime(df['order_date'],format='%Y-%m-%d')
```

```python
# Drop list_price and  discount_percent

# df.dtypes
df.drop(columns=['cost_price','list_price','discount_percent'],inplace=True)
```
